```python
import pandas as pd

main_df = pd.read_csv(r'C:\Users\akozhevnikov\Downloads\hakatone\bsm_data_train.csv')
main_df = main_df.drop_duplicates('ident_bag', keep='first').reset_index(drop=True)
main_df
```

```python
main_df['MessageReceivedSeconds'] = pd.to_datetime(main_df['MessageReceivedDate']).apply(lambda x: x.value) / 1000000000
main_df['MessageReceivedSeconds'] -= main_df['MessageReceivedSeconds'].min()
main_df['MessageReceivedSeconds'] = (main_df['MessageReceivedSeconds'] // 60).astype(int)
main_df = main_df.sort_values(by='MessageReceivedSeconds')
main_df
```

```python
main_df['Recived_Processed'] = (pd.to_datetime(main_df['MessageProcessedDate']) - pd.to_datetime(main_df['MessageReceivedDate'])).apply(lambda x: x.seconds)
```

```python
%%time
df = main_df[['MessageReceivedSeconds', 'Recived_Processed']].groupby('MessageReceivedSeconds').sum()
```

```python
def calculate_anomaly(new_values: pd.Series, # список значений новых
                      column_name: str, # какой признак, должен быть в stats
                      stats: dict, 
                      update_stats=True):
    result = (new_values - stats[column_name]['mean']) >  stats[column_name]['std']*3
    iqr = stats[column_name]['q75'] - stats[column_name]['q25']
    lower_bound = stats[column_name]['q25'] - 1.5 * iqr
    upper_bound = stats[column_name]['q75'] + 1.5 * iqr
    result += new_values > upper_bound
    if lower_bound > 0:
        result += new_values < lower_bound
    return result
```

```python
# исторические базовые данные 
# TODO сделать чтобы они обновлялись 
stats = {'Recived_Processed': {'mean': df['Recived_Processed'].mean(), 
                               'count': len(df['Recived_Processed']),
                              'std': df['Recived_Processed'].std(),
                              'q25': df['Recived_Processed'].quantile(0.25),
                               'q50': df['Recived_Processed'].quantile(0.50),
                               'q75': df['Recived_Processed'].quantile(0.75)}}
```

```python
new_values = pd.Series([38., 1000., 40., 35., 99.])
```

```python
result = calculate_anomaly(new_values, 'Recived_Processed', stats)
```
